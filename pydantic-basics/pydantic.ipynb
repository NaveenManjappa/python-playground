```python
from pydantic import BaseModel,EmailStr,ValidationError
```

```python
class UserInput(BaseModel):
  name:str
  email: EmailStr
  query:str
```

```python
user_input = UserInput(
  name="John Doe",
  email="John.Doe@gmail.com",
  query="I have forgotten my password"
)
```

```python
print(user_input)
```

```python
user_input = UserInput(
  name="12",
  email="John.Doe@gmailcom",
  query="I have forgotten my password"
)
print(user_input)
```

```python
def validate_user_input(input_data):
  try:
      user_input = UserInput(**input_data)
      print(f"Valid user input created:")
      print(f"{user_input.model_dump_json(indent=2)}")
      return user_input
  except ValidationError as e:
      print("Validation error occurred")
      #print(e.errors())
      for error in e.errors():
         print(f"{error['loc'][0]}: {error['msg']}")
      return None
```

```python
input_data = {
  "name":12,
  "email":"John.Doegmail.com",
  "query":"I have forgotten password!",
  "orderId":"122"
}

user_input = validate_user_input(input_data)
print(user_input)
```

```python
# Attempt to create an instance of UserInput with missing query field
input_data = {
  "name":"John Doe",
  "email":"John.Doe@gmail.com" 
}

user_input = validate_user_input(input_data)
print(user_input)
```

```python
from pydantic import Field
from typing import Optional
from datetime import date

#Define a new UserInput model with optional fields
class UserInput(BaseModel):
  name:str
  email:EmailStr
  query:str
  order_id: Optional[int] = Field(
    None,
    description="5-digit order number(cannot start with 0)",
    le=99999,
    ge=10000
  )
  purchase_date:Optional[date]=None



```

```python
#Define a dictionary with required fields only
input_data = {
  "name":"Joe Doe",
  "email":"joe.doe@gmail.com",
  "query":"I have forgotten my password"
}

user_input = validate_user_input(input_data)
```

```python
print(user_input)
```

```python
#Define a dictionary with required fields only
input_data = {
  "name":"Joe Doe",
  "email":"joe.doe@gmail.com",
  "query":"I have forgotten my password",
  "order_id":1234,
  "purchase_date":date(2025,12,25)
}

user_input = validate_user_input(input_data)
```

```python
#Define a dictionary with all fields and some additional fields which are not there in the model
input_data = {
  "name":"Joe Doe",
  "email":"joe.doe@gmail.com",
  "query":"I have forgotten my password",
  "order_id":12325,
  "purchase_date":"2025-12-25",
  "system_message":"Logging status regarding the order processing",
  "iteration":1
}

user_input = validate_user_input(input_data)
print(user_input)
```
